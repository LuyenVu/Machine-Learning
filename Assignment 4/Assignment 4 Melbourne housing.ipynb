# Cleaning data

# 
Column 'Distance': 1 row, Column 'Postcode': 1 row.

```python
data[data.Distance.isnull()]
data[data.Postcode.isnull()]
```

# 
Null value of Distance and Postcode is same row.
Because it have 'Price' value, 'Address', 'Room' info and 'Price' feature is important so it will assign 'Distance' to 0 (zero), 'Postcode' to empty.

```python
data.Distance.fillna(0, inplace=True)
data.Postcode.fillna("", inplace=True)
```

# 
Columns: 'CouncilArea', 'Regionname', 'Propertycount' these is same rows. It have info as: price, Rooms, Suburb, Method ...need to analyst.

```python
# Run line by line to check it
data[data.CouncilArea.isnull()]
data[data.Regionname.isnull()]
data[data.Propertycount.isnull()]
```

```python
data.CouncilArea.fillna("", inplace=True)
data.Regionname.fillna("", inplace=True)
data.Propertycount.fillna(data.Propertycount.mean(), inplace=True)
```

# 
Column 'YearBuild' data: we check data with logic and actualy.

```python
data.YearBuilt[data.YearBuilt > 2019].value_counts(dropna=False)
```

```python
data.YearBuilt.replace(2106, 2016, inplace=True)
```

```python
data.YearBuilt[data.YearBuilt < 1800].value_counts(dropna=False)
```

```python
data.YearBuilt.replace(1196, 1961, inplace=True)
```

# 
'BuildingArea','Lattitude', 'Longtitude','Landsize','Car' assign to zero value 

```python
# Run line by line to view dataframe.
data[data.BuildingArea.isnull()]
data[data.BuildingArea.isnull()]
data[data.BuildingArea.isnull()]
```

```python
# Run command line by line to update data
data.BuildingArea.fillna(0, inplace=True)
data.Lattitude.fillna(0, inplace=True)
data.Longtitude.fillna(0, inplace=True)
data.Landsize.fillna(0, inplace=True)
data.Car.fillna(0, inplace=True)
```

# 2.Apply the model for prediction with holdout and cross validation


```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

# 2.1 Holdout

```python
dataPriced.head()
```

```python
dataPriced.info()
```

```python
# 'Postcode' is object type 
# get float64, int type dataframe for regression
dataRegr = dataPriced[['Rooms','Price','Distance','Car','Lattitude','Longtitude','Landsize','BuildingArea','Propertycount']] 
```

```python
ligres=linear_model.LinearRegression()
```

```python
train, test = train_test_split(dataRegr, test_size = 0.2, random_state=512)
```

```python
train.shape
```

```python
test.shape
```

```python
X_train = train.loc[:, dataRegr.columns != 'Price']
y_train = train.Price

X_test = test.loc[:, dataRegr.columns != 'Price']
y_test = test.Price
```

```python
ligres.fit(X_train.values, y_train.values)
```

```python
predict_train = ligres.predict(X_train.values)
print("Mean squared error is:",mean_squared_error(y_train, predict_train))
```

```python
predict_test = ligres.predict(X_test.values)
print("Mean squared error is:",mean_squared_error(y_test, predict_test))
```

```python
fig, ax = plt.subplots()
ax.scatter(y_test, predict_test)
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```

# 2.2 Cross validation

```python
from sklearn.model_selection import cross_val_predict
```

```python
ligres_crosVal = linear_model.LinearRegression()
```

```python
y = dataRegr.Price
X = dataRegr.loc[:, dataRegr.columns != 'Price']
predicted = cross_val_predict(ligres_crosVal, X.values, y.values, cv=5)

fig, ax = plt.subplots()
ax.scatter(y, predicted)
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```
